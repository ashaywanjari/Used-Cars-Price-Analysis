```python
#clean Price
cars["Price"] = (
    cars["Price"].str.replace("₹", "", regex=False)
    .str.replace("Lakh", "", regex=False)
    .str.strip()
)

cars["Price"] = pd.to_numeric(cars["Price"], errors="coerce") * 100000

```

```python
#clean km_driven
cars['KM_driven'] = (cars['KM_driven'].str.replace("kms","",regex=False)
                     .str.replace(",","",regex=False)
                     .str.strip()
                    )
cars['KM_driven']=pd.to_numeric(cars['KM_driven'], errors="coerce")
```

```python
# Year
cars['Year_of_Manufacture'] = pd.to_numeric(cars['Year_of_Manufacture'], errors = "coerce")
```

```python
cars.isna().sum()
```

```python
cars["Price"].fillna(cars["Price"].median(), inplace=True)
```

```python
# Fill missing values
cars['Fuel_Type'] = cars['Fuel_Type'].replace('•', np.nan)
cars["Fuel_Type"].fillna(cars["Fuel_Type"].mode()[0], inplace=True)
```

```python
cars.isna().sum()
```

```python
cars.duplicated().sum()
```

```python
cars = cars.drop_duplicates()
cars
```

```python
# Feature Engineering 
current_year = datetime.datetime.now().year
#car_age 
cars['car_age'] = current_year - cars['Year_of_Manufacture']
cars['car_age'] = cars['car_age'].fillna(0)
#kms_per_year
cars['km_per_year'] = cars['KM_driven']/cars['car_age']
cars['km_per_year'] = round(cars['km_per_year'],2)


# cars segments 
import numpy as np

cars["price_segment"] = pd.cut(
    cars["Price"],
    bins=[0, 300000, 700000, 2000000, np.inf],
    labels=["Budget", "Mid", "Premium", "Luxury"]
)


```

```python
#city wise car count 

car = cars['City'].value_counts().index

plt.figure(figsize=(10,7))
sns.countplot(y='City', data=cars, order=car)
plt.title('City Wise cars distribution')
plt.xlabel("No_of_cars")
plt.ylabel('City')
plt.show()
```

```python
# Average Price per city 
Price = cars.groupby('City')['Price'].mean().index


ax = sns.barplot(x='Price', y='City', data=cars, order=Price)

# Add labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

plt.show()
```

```python
fuel = cars['Fuel_Type'].value_counts()
plt.bar(fuel.index, fuel.values, color = "GREEN", width = 0.8)
plt.xlabel("Fuel Type")
plt.ylabel("COUNT")
plt.title("Fuel Type Distribution")
plt.show()
```

```python
# KM Driven VS Price 
bins = [ 0, 20000, 40000, 60000, 80000, 100000, 200000]
labels = ['0-20k', '20-40k', '40-60k', '60-80k','80-100k', '100+']

#temporary bins (no column added)  
km_bins = pd.cut(cars['KM_driven'], bins = bins, labels=labels)

#Average Price per bin bin
bin_price = cars.groupby(km_bins)['Price'].mean()

plt.figure(figsize = (10,5))
plt.scatter(bin_price.index.astype(str), bin_price.values)
plt.plot(bin_price.index.astype(str), bin_price.values, marker='o')
plt.xlabel('kilometers Driven')
plt.ylabel('Average Price')
plt.title('KM Driven VS Average Price')
plt.xticks(rotation=45)
plt.show()
```
