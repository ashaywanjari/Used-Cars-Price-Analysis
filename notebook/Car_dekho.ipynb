##### Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

##### HTTP Request

```python
# store website in variables
website = 'https://www.cardekho.com/used-maruti+cars+in+new-delhi'
```

###### Get Request and Status Code

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
    "Referer":'https://www.cardekho.com/used-maruti+cars+in+new-delhi'
}

try:
    # Send a GET request with headers
    response = requests.get(website, headers=headers)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        print("Access Granted. Status Code:", response.status_code)
    else:
        print("Unexpected Status Code:", response.status_code)

except requests.exceptions.RequestException as e:
    print("Error:", e)
```

###### Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
 # creating empty list and inside in for loop
name = [] 
model = []
year = [] 
fuel_type = [] 
km_driven = [] 
transmission = [] 
price = []


for result in results:
    #name
    try:
        name.append(result.find('h3').get_text().split()[1])
    except:
        name.append('n/a')
    
    #model
    try:
        m_name = result.find('h3').get_text().split()[2:6]
        model_name = " ".join(m_name)
        model.append(model_name)
        
    except:
        model.append('n/a')


    #year
    try:
        year.append(result.find('h3').get_text().split()[0])
    except:
        year.append('n/a')

    #fuel_type
    try:
        fuel_type.append(result.find('div',{'class':'dotsDetails'}).get_text().split()[3])
    except:
        fuel_type.append('n/a')

    #km_driven 
    try:
        km = result.find('div',{'class':'dotsDetails'}).get_text().split()[0:2]
        km_d = " ".join(km)
        km_driven.append(km_d)
    except:
        km_driven.append('n/a')

    #transmission
    try:
        transmission.append(result.find('div',{'class':'dotsDetails'}).get_text().split()[-1])
    except:
        transmission.append('n/a')

    #price
    try:
        price.append(result.find('div',{'class':'Price hover'}).get_text())
    except:
        price.append('n/a')

    
    
```

###### Create Pandas Dataframe

```python
car_dekho = pd.DataFrame({'Name':name, 'Model':model, 'Year_of_Manufacture':year,
                          'Fuel_Type':fuel_type, 'KM_driven':km_driven, 'Price':price})
```

```python
city = []
name = [] 
model = []
year = [] 
fuel_type = [] 
km_driven = [] 
transmission = [] 
price = []

cities = ["Ahmedabad", "Bangalore", "Chandigarh", "Chennai", "Delhi NCR",
               "Gurgaon", "Hyderabad", "Jaipur", "Mumbai", "Noida", "Pune", "Nagpur"]


brands = ["Maruti", "Hyundai", "Honda", "Toyota", "Mahindra", "Tata", "Mercedes-Benz", "BMW", "Audi",
         "Land Rover", "Kia", "MG", "Ford", "Renault", "Volkswagen", "Nissan", "Datsun"]

for city_names in cities:
    for brand in brands:
    #website in variable
        for page in range(1,20):
        #website in variable
            website = website = f"https://www.cardekho.com/used-{brand}+cars+in+{city_names}/page-{page}"

            #request to website
            headers = {
             "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
             "Referer":f"https://www.cardekho.com/used-{brand}+cars+in+{city_names}/page-{page}"
}
            try:
             #Send a GET request with headers
                response = requests.get(website, headers=headers)
           # Check if the request was successful (status code 200)
                if response.status_code == 200:
                    print("Access Granted. Status Code:", response.status_code)
                else:
                    print("Unexpected Status Code:", response.status_code)
            except requests.exceptions.RequestException as e:
                print("Error:", e)

            #soup objects
            soup = BeautifulSoup(response.content, 'html.parser')

           #result
            results = soup.find_all('div',{'class':'NewUcExCard posR'})

            for result in results:

                #city
                city.append(city_names)
                
                
                
                #name
                try:
                    name.append(result.find('h3').get_text().split()[1])
                except:
                    name.append('n/a')

                #model
                try:
                    m_name = result.find('h3').get_text().split()[2:6]
                    model_name = " ".join(m_name)
                    model.append(model_name)
        
                except:
                    model.append('n/a')
     
                #year
                try:
                    year.append(result.find('h3').get_text().split()[0])
                except:
                    year.append('n/a')

                #fuel_type
                try:
                    fuel_type.append(result.find('div',{'class':'dotsDetails'}).get_text().split()[3])
                except:
                    fuel_type.append('n/a')

                #km_driven 
                try:
                    km = result.find('div',{'class':'dotsDetails'}).get_text().split()[0:2]
                    km_d = " ".join(km)
                    km_driven.append(km_d)
                except:
                    km_driven.append('n/a')

                #transmission
                try:
                    transmission.append(result.find('div',{'class':'dotsDetails'}).get_text().split()[-1])
                except:
                    transmission.append('n/a')

                #price
                try:
                    price.append(result.find('div',{'class':'Price hover'}).get_text())
                except:
                    price.append('n/a')
```

```python
car_dekho = pd.DataFrame({'City':city, 'Name':name, 'Model':model, 'Year_of_Manufacture':year,
                          'Fuel_Type':fuel_type, 'KM_driven':km_driven,
                          'Transmission':transmission, 'Price':price})
```
